```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
w = 4   # UDL in N/m
L_total = 4  # Length of the beam
a = 1
b = 2
c = 1

# Reactions at supports (assuming a simply supported beam)
R1 = w * b * (c + b/2) / L_total
R2 = w * b * (a + b/2) / L_total

# Discretize the beam
X = np.linspace(0, L_total, 200)
SF = []
M  = []

for x in X:
    if x < a:
        sf = R1
        m = R1 * x
    elif a <= x <= (a+b):
        sf = R1 - w*(x-a)
        m = R1*x - w*(x-a)**2/2
    else:
        sf = -R2
        m = R2 * (L_total - x)
    SF.append(sf)
    M.append(m)

# Plotting
plt.figure(figsize=(7.5,10))

plt.subplot(2,1,1)
plt.plot(X, SF, color='green')
plt.fill_between(X, SF, color='green', hatch='||', alpha=0.5)
plt.title("Shear Force Diagram")
plt.xlabel("Length of Beam (m)")
plt.ylabel("Shear Force (N)")
plt.xlim(0, L_total)

plt.subplot(2,1,2)
plt.plot(X, M, color='red')
plt.fill_between(X, M, color='red', hatch='\\', alpha=0.5)
plt.title("Bending Moment Diagram")
plt.xlabel("Length of Beam (m)")
plt.ylabel("Moment (Nm)")

plt.tight_layout(pad=4.0)
plt.show()
```
