```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin, cos, pi

syt = float(input(" enter yeild strength of material : "))
s1  = float(input(" enter stress in direction σ1 : "))
s2  = float(input(" enter stress in direction σ2 : "))

sv  = np.sqrt(s1**2 - (s1*s2) + s2**2)

# Ellipse parameters
a = 2**0.5 * syt      # semi major axis
b = (2/3)**0.5 * syt  # semi minor axis

theta = np.linspace(0, 2*pi, 720)  # 0 to 2π radians

x = a * sin(theta)
y = b * cos(theta)

 
r_angle = pi/4
xd = x * cos(r_angle) - y * sin(r_angle)
yd = x * sin(r_angle) + y * cos(r_angle)

# Scatter points
sp1 = [232, 23, 123, 90]
sp2 = [-90, 32, 213, 123]

# Max shear region
xs = [syt, syt, 0, -syt, -syt, 0, syt]
ys = [0, syt, syt, 0, -syt, -syt, 0]

plt.figure(figsize=(6,6))
plt.plot(xd, yd, label="Von Mises Yield Ellipse")
plt.plot(xs, ys, linestyle='--', label="Max Shear Region")
plt.scatter(sp1, sp2, color='red', label="Stress Points")

# Add dotted axes
plt.axhline(0, color='black', linestyle='--')  # x-axis
plt.axvline(0, color='black', linestyle='--')  # y-axis

plt.xlabel("σ₁ (Stress 1)")
plt.ylabel("σ₂ (Stress 2)")
plt.title("Von Mises Yield Plot")
plt.legend(loc='upper left',fontsize = 8)  

plt.show()
```
